```python
import requests
import json
import pandas as pd

# Import the Google Maps API client library
from googlemaps import Client

# Set the API key
API_KEY = "AIzaSyBxCevn29vIjZ-yDlyeW4ggguX5d_pcSh4"

# Create a GooglePlaces object
client = Client(API_KEY)

# Set the location to Nashville, TN
location = "Nashville, TN"

# Search for grocery stores near Nashville
response = client.places_nearby(type="grocery_store", location=location)

# Get the results
results = response["results"]

# Create a pandas dataframe
df = pd.DataFrame(results)

# Print the dataframe
df
```
