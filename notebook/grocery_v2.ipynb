```python
grocery_df = pd.concat([publix_df, kroger_df, foodlion_df, target_df, walmart_df, piggly_df, foodcity_df, aldi_df, dgeneral_df, familyd_df, freshmarket_df, sprouts_df, trader_df, wholefoods_df], axis=0)

grocery_df.head()
```

```python
grocery_df = grocery_df.drop(['icon', 'icon_background_color', 'icon_mask_base_uri', 'opening_hours', 'photos', 'plus_code', 'rating', 'reference', 'types', "user_ratings_total", 'price_level', 'permanently_closed'], axis=1)
grocery_df
```

```python
grocery_df = grocery_df[grocery_df['formatted_address'].str.contains('Nashville')]
grocery_df = grocery_df.reset_index(drop=True)
grocery_df
```

```python
grocery_df = grocery_df[grocery_df['business_status'].str.contains('OPERATIONAL')]
grocery_df = grocery_df.reset_index(drop=True)
grocery_df
```

```python
grocery_df = grocery_df.drop_duplicates(subset='formatted_address')
grocery_df = grocery_df.reset_index(drop=True)
grocery_df
```

```python
grocery_df['zipcode'] = grocery_df['formatted_address'].str.extract(r'(\b\d{5}\b)')
grocery_df
```

```python
def extract_coordinates(row):
    return row['location']['lat'], row['location']['lng']
grocery_df[['lat', 'lng']] = grocery_df['geometry'].apply(extract_coordinates).apply(pd.Series)
grocery_df = grocery_df.drop('geometry', axis=1)

grocery_df
```

```python
income_df = pd.read_csv('../data/income.csv')
```

```python
income_df['Zip code'] = income_df['Zip code'].astype(str)
merged_df = grocery_df.merge(income_df, left_on='zipcode', right_on='Zip code', how='left')  
grocery_df['average income'] = merged_df['Average Income']
grocery_df
```

```python
folder_path = '../data/'
grocery_df.to_csv(folder_path + 'grocery.csv', index=False)
```
