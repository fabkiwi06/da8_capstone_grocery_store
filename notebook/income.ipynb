```python
#Data From https://simplemaps.com/city/nashville/zips/income-household-median

import pandas as pd
data = {
    'Zip code': ['37013', '37027', '37072', '37076', '37080', '37115', '37138', '37143', '37189', '37201',
                 '37203', '37204', '37205', '37206', '37207', '37208', '37209', '37210', '37211', '37212',
                 '37214', '37215', '37216', '37217', '37218', '37219', '37220', '37221', '37228'],
    'Average Income': [59015, 139932, 63784, 65870, 56763, 46123, 72861, 69934, 75690, 76014, 57330, 87292, 106241,
                       59203, 42973, 39964, 54461, 36283, 56748, 63586, 61558, 127203, 61641, 49103, 46974, 87344,
                       119803, 83331, 31451]
}

income_df = pd.DataFrame(data)
income_df
```

```python
income_ranges = {
    'Under 30k': (0, 29999),
    '30k-39k': (30000, 39999),
    '40k-49k': (40000, 49999),
    '50k-59k': (50000, 59999),
    '60k-69k': (60000, 69999),
    '70k-79k': (70000, 79999),
    '80k-89k': (80000, 89999),
    '90k-99k': (90000, 99999),
    '100k+': (100000, float('inf'))
    }

def map_income_range(income):
    for range_name, (range_start, range_end) in income_ranges.items():
        if range_start <= income <= range_end:
            return range_name
    return ""    

income_df['Income Range'] = income_df['Average Income'].apply(map_income_range)

income_df
```

```python
folder_path = '../data/'
income_df.to_csv(folder_path + 'income.csv', index=False)
```
